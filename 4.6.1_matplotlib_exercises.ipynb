```python
# 2. Use matplotlib to plot the following equation. Add an anotation for the point 0, 0, the origin.

x = list(range(-49, 50))
y = [((n ** 2)-n+2) for n in x]
plt.scatter(x, y)
plt.title('An Exponential Distribution, $y = x^2$ -x +2')
plt.annotate('Origin', xy=(0, 0), xytext=(0, 300),
             arrowprops={'facecolor': 'blue'})
plt.xlabel('$x$')
plt.ylabel('$x^2$')
plt.show()
```

```python
# 3. Create and label 4 separate plots for the following equations (choose a range for x that makes sense). 
# You can use functions from the math module to help implement some of the equations above.

import math
```

```python
#first equation y=√x

x = list(range(0, 100))
y = [(n ** 0.5) for n in x]
plt.plot(x,y)
plt.title('y=√x')
plt.show()
```

```python
#second equation y=x^3

x = list(range(-10, 11))
z = [(n** 3) for n in x]
plt.plot(x,z)
plt.title('y=x^3')
plt.show()
```

```python
#third equation y=tan⁡(x)

x = list(range(-5, 6))
q = [math.atan(n) for n in x]
plt.plot(x,q)
plt.title('y=tan⁡(x)')
plt.show()
```

```python
#fourth equation y=2^x

x = list(range(-5, 6))
p = [(2 ** n) for n in x]
plt.plot(x,p)
plt.title('y=2^x')
plt.show()
```

```python
# 4. Combine the figures you created in the last step into one large figure with 4 subplots.

import math

n_rows = 2
n_cols = 2

x = list(range(-5, 6))
y = [(n ** 0.5) for n in x]
z = [(n** 3) for n in x]
q = [math.atan(n) for n in x]
p = [(2 ** n) for n in x]

plt.figure(figsize=(6, 6)).subplots_adjust(hspace=0.5) # (width, height)


# the first subplot
plt.subplot(n_rows, n_cols, 1)
plt.plot(x,y)
plt.title('Sqrt of x')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, z)
plt.title('x Cubed')

# the third subplot
plt.subplot(n_rows, n_cols, 3)
plt.plot(x, q)
plt.title('tangent of x')

# the fourth subplot
plt.subplot(n_rows, n_cols, 4)
plt.plot(x, p)
plt.title('Two to the power of x')

plt.show()

```

```python
# 5. Comine the figures you created in the last step into one figure where 
# each of the 4 equations has a different color for the points. Be sure to include a legend.


# 4. Combine the figures you created in the last step into one large figure with 4 subplots.

import math

x = list(range(-10, 10))
y = [(n ** 0.5) for n in x]
z = [(n** 3) for n in x]
q = [math.atan(n) for n in x]
p = [(2 ** n) for n in x]

plt.figure(figsize=(6, 6))


plt.plot(x, y, c='green', alpha=0.8, label='sqrt of x')
plt.plot(x, z, c='red', alpha=0.4, label='x cubed')
plt.plot(x, q, c='purple', alpha=0.4, label='tangent of x')
plt.plot(x, p, c='blue', alpha=0.4, label='2 to the power of x')
plt.ylim(-10, 10)


plt.legend(loc='best')

plt.show()


```
